```python
def simple_mid_price(row):
    bid = row[3]
    ask = row[9]
    return (bid+ask) / 2
```

```python
def generate_training_for_future_price_pred(base_series, target_series, iters=0, lag=0):
    # create array with data shifted by 'iters' into the future

    # return in
    if iters > 0:
        
        base_present_data = base_series[:len(base_series)-iters]
        base_future_data = base_series[iters:]

        base_series = base_future_data - base_present_data

        target_present_data = target_series[:len(target_series)-iters]
        target_future_data = target_series[iters:]

        target_series = target_future_data - target_present_data
        
    
    if lag != 0:
        if lag < 0:
            # negative lag means we need to shift the target series to the left
            target_series = target_series[:len(target_series)+lag]
        else:
            # positive lag means shift target to the right
            target_series = target_series[lag:]

        base_series = base_series[:len(base_series)-abs(lag)]
    
    return base_series, target_series
```

```python
def plot_target_vs_base(target_day, base_day, target_data_name, base_data_name, r_squared, using_last_year=True, use_iters=0, use_lags=0):
    original_target_day = target_day
    original_base_day = base_day
    original_target_name = target_data_name
    original_base_name = base_data_name

    if target_day == '-1': target_day = 'neg1'
    elif target_day == '-2': target_day = 'neg2'
    if base_day == '-1': base_day = 'neg1'
    elif base_day == '-2': base_day = 'neg2'


    target_name = target_data_name + '_' + target_day
    base_name = base_data_name + '_' + base_day

    if using_last_year:
        target_series = this_year_data[this_year_names.index(target_name)]
        base_series = previous_year_data[last_year_names.index(base_name)]
    else:
        target_series = this_year_data[this_year_names.index(target_name)]
        base_series = this_year_data[this_year_names.index(base_name)]

    base_series, target_series = generate_training_for_future_price_pred(base_series, target_series, iters=use_iters, lag=use_lags)
            
    plt.plot(target_series, label=target_name, c='red')
    plt.twinx()
    plt.plot(base_series, label=base_name, c='blue')
    plt.title(f'{original_target_name} ({original_target_day}) vs {original_base_name} ({original_base_day}) (r_squared: {r_squared:.2f})')
    plt.legend()


def plot_from_df(df, row, use_last_year=True, use_iters=0, use_lag=0):
    row_to_plot = df.iloc[row]
    target_day = row_to_plot['target_day']
    target_name = row_to_plot['target_name']
    base_name = row_to_plot['base_name']
    base_day = row_to_plot['base_day']
    r_squared = row_to_plot['r_squared']
    plot_target_vs_base(target_day, base_day, target_name, base_name, r_squared, use_last_year, use_iters, use_lag)
```
